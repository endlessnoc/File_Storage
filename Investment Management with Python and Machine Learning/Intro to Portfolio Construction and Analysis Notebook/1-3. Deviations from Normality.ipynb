# Deviations from Normality

```python
import pandas as pd
import numpy as np
import scipy.stats
import risk_kit as rk
%load_ext autoreload
%autoreload 2
```

```python
hfi = rk.get_hfi_returns()
hfi.head()
```

```python
# To check if returns are negative-skewed.
# Mean > Median means that negative-skewed
pd.concat([hfi.mean(), hfi.median(), hfi.mean() > hfi.median()], axis= 'columns')
```

```python
#Skewness
#If the distribtion is normal, we will get a skewness of 0
rk.skewness(hfi).sort_values()
```

```python
hfi.shape
```

```python
#Generate a random normal numbers and check the skewness
normal_rets= np.random.normal(loc= 0, scale= 0.15, size=(263,1)) #loc= mean, scale= S.D.
rk.skewness(normal_rets)
```

```python
#If we increase # of values then skewness will definely more closely to 0
normal_rets2= np.random.normal(loc= 0, scale= 0.15, size=(263000,1)) #loc= mean, scale= S.D.
rk.skewness(normal_rets2)
```

```python
#Kurtosis
rk.kurtosis(hfi)
```

```python
#The normally generated returns should have a kurtosis of 3
rk.kurtosis(normal_rets)
```

If we use scipy.stats.kurtosis instead, it'll give us a kurtosis approx. 0, since it's giving us the excess kurtosis over 3.   
Here we use our own code(not subtracting 3) so it's approx. 3

```python
rk.kurtosis(normal_rets2)
```

we know that the returns for these hedge fund indices arent actually symmetrically distributed since they seem to have negative skewness in most cases and positive skewness in some other cases.

```python
#Applying test on normal_rets Series
#We can say with a high degree of confidence that normal_rets is normally distributed since p-value is high.
scipy.stats.jarque_bera(normal_rets)
```

```python
rk.is_normal(normal_rets)
```

```python
#df.aggregate(func) can apply function over the specified axis, here we apply rk.is_normal() on every columns 
hfi.aggregate(rk.is_normal)
```

```python
#Lets test that the other developed code to check if the Small&Large Cap returns are normally distributed or not. 
ffme = rk.get_ffme_returns()
rk.skewness(ffme)
```

```python
rk.kurtosis(ffme)
```

We get enormous kurtosis.
