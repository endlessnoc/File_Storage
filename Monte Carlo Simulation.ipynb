When it comes to Monte Carlo simulations, a classic example is to estimate the value of Ï€(pi).   
The Monte Carlo method involves generating random points within a shape(E.g. Square), and then calculating the proportion of these points that fall within another shape(E.g. Circle), aiming to estimate the area ratio of the shapes.

```python
import random
import matplotlib.pyplot as plt

Num_simulate = 1000

Num_inside_circle = 0
Num_outside_circle =  0


#Setting the coordinate(x,y) of points in the cicle
Circle_x = []
Circle_y = []


for simulate in range(Num_simulate):
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    Distance = (x**2 + y**2)  
    if Distance <= 1:
        Num_inside_circle = Num_inside_circle + 1
        Circle_x.append(x)
        Circle_y.append(y)
    else:
        Num_outside_circle = Num_outside_circle + 1

# Calculate the vluae of Ï€(pi), the explanation of the caculation method is stated at the end. 
Pi_estimate = (Num_inside_circle / Num_simulate) * 4


plt.figure(figsize=(6, 6))
plt.title(f"Monte Carlo Simulation of Ï€ (Samples size: {Num_simulate})")
plt.scatter(Circle_x, Circle_y, color="DarkViolet", marker="o")
plt.scatter([], [], color="white", label=f"Ï€ â‰ˆ {Pi_estimate:.5f}")
plt.legend()
plt.show()


```

## The explanation of the calculation 
We randomly put some points in a square and calculate the ratio of points in the circle to total points (Noted that the circle is in the square). This ratio, scaled by 4, provides an estimate of pi because the circle occupies a quarter of the square's area.   
The more points we use (Num_simulation), the closer our estimation gets to the value of pi (Ï€).
