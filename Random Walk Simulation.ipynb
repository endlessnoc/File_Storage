# Random Walk Simulation

## 1D

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Setting Seed to ensure simulation results are replicable.
# np.random.seed(1)

Num_steps = 10000
Step_size = 1
Random_steps = np.random.choice([-1, 1], size=Num_steps) * Step_size
RW = np.cumsum(Random_steps)


fig, ax = plt.subplots(figsize=(10,6))
plt.title("Random Walk Simulation", fontsize=20)
plt.plot(RW)
plt.xlabel("Steps")
plt.ylabel("Position")
plt.grid()
plt.show()
```

```python
RW_Data = pd.DataFrame(RW, columns = ["Position"])
RW_Data.index = range(1, Num_steps+1)
RW_Data
```

## 2D Simulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Setting Seed to ensure simulation results are replicable.
# np.random.seed(42)

Num_steps = 100000
Step_size = 1

Random_steps_X = np.random.choice([-1, 1], size=Num_steps) * Step_size
Random_steps_Y = np.random.choice([-1, 1], size=Num_steps) * Step_size
RW_X, RW_Y  = np.cumsum(Random_steps_X), np.cumsum(Random_steps_Y)

fig, ax = plt.subplots(figsize=(10,6))
plt.title("2D Random Walk Simulation with End Point", fontsize=20)
plt.plot(RW_X, RW_Y)
plt.scatter(RW_X[-1], RW_Y[-1], color='red', label='End Point',zorder=3)
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.legend()
plt.grid()
plt.show()
```

```python
RW_Data = pd.DataFrame({"X": RW_X,
                        "Y": RW_Y})
RW_Data.index = range(1,Num_steps+1)
RW_Data
```
